```python
import pandas as pd
df = pd.read_csv("Extracted.csv")
print(df.head())
```

```python
#finding Most expensive product and Cheapest Product 

max = df['Price'][0]
min = df['Price'][0]
for i in df['Price']:
    max = i if i > max else max
    min = i if i < min else min  
print(df.loc[df['Price'] == max])
print(df.loc[df['Price'] == min])


print("===============================================================================")

#Also sorting the data based on Price valeue

Sorted = df.sort_values('Price')
Sorted
```

```python
# for referencing
#brand = len(df.loc[df['Brand'] == "Generic"])
#print(brand)

#Making  a list of uniqe Brand
Brands_name  = []
for i in df['Brand']:
    if i not in Brands_name:
        Brands_name.append(i)
for i in Brands_name:
    print(i)
```

```python
# Number of Products from Each brand in dictionry

NoOfProductForEachBrand = {}
for i in Brands_name:
    No_products = len(df.loc[df['Brand'] == i])
    NoOfProductForEachBrand[i] = No_products
print(NoOfProductForEachBrand)
```

```python
# print the brand name and no of products each brand

for k, v in NoOfProductForEachBrand.items():
    print(f'{k} have {v} products')
```
